```python
# multiplication table
def print_multiplication_table():
  size = int(input("Enter the size of the multiplication table (maximum 18 :"))
  if size > 18:
     print("size is too large.Please enter a value less than or equal to 18.")
     return

  # print the top row of the table
  print(" |", end="")
  for i in range(1,size+1):
    print(f" {i:2} ", end="")
  print()
  print(" " + " "*size)
  # print the rest of the table
  for i in range(1,size+1):
    print(f"{i:2}|", end="")
    for j in range(1, size+1):
      print(f" {i*j:2} ", end="")
    print()
print_multiplication_table()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

# create example DataFrame
df = pd.DataFrame({
    "Class": ["A", "B", "A", "B", "C", "C"],
    "Game": ["X", "X", "Y", "Y", "X", "Y"]
})

# count occurrences of each unique combination of "Class" and "Game"
g = df.groupby(["Class", "Game"]).size()

# extract x and y values from g for plotting
x = range(len(g))
y = g.values

# create bar chart
plt.bar(x, y)

# set tick labels
plt.xticks(x, g.index, rotation=45)

# set axis labels
plt.xlabel("Class/Game")
plt.ylabel("Count")

# show plot
plt.show()
```
